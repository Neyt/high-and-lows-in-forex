```python
from datetime import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = url
#data = "GBPUSD_Candlestick_15_M_BID_01.08.2013-01.09.2018.csv"
```

```python
# Utility functions

def find_max_indices(a):
    inx = list(range(a.size))
    zipped = list(zip(a,inx))
    
    filtered = []
    for i in list(range(a.size -1)):
        if abs(zipped[i+1][0] - zipped[i][0]) > 0.0000001:
            filtered.append(zipped[i])
    max_index = []
    for i in list(range(len(filtered) - 4)):
        if filtered[i+2][0] > filtered[i][0] and filtered[i+2][0] > filtered[i+1][0] and filtered[i+2][0] > filtered[i+3][0] and filtered[i+2][0] > filtered[i+4][0]:
            max_index.append(filtered[i+2][1])
    
    return max_index

def find_min_indices(a):
    inx = list(range(a.size))
    zipped = list(zip(a,inx))
    
    filtered = []
    for i in list(range(a.size -1)):
        if abs(zipped[i+1][0] - zipped[i][0]) > 0.0000001:
            filtered.append(zipped[i])
    min_index = []
    for i in list(range(len(filtered) - 4)):
        if filtered[i+2][0] < filtered[i][0] and filtered[i+2][0] < filtered[i+1][0] and filtered[i+2][0] < filtered[i+3][0] and filtered[i+2][0] < filtered[i+4][0]:
            min_index.append(filtered[i+2][1])
    
    return min_index

def sparse_to_boolean(sparse, size):
    sparse_pos = set(sparse)
    boolean = []
    for i in list(range(size)):
        if i in sparse_pos:
            boolean.append(True)
        else:
            boolean.append(False)
    return np.array(boolean, dtype=np.bool)

def tops(a):
    return sparse_to_boolean(find_max_indices(a), a.size)

def bottoms(a):
    return sparse_to_boolean(find_min_indices(a), a.size)
```

```python
# Test for utility functions
tops(np.array([10,2,1,4,5,6,6,7,34,2,5,25,3,-26,-20,-29],dtype=np.float))
```

```python
# Test for utility functions
bottoms(np.array([10,2,1,4,5,6,6,7,34,2,5,25,3,-26,-20,-29],dtype=np.float))
```

```python
df = pd.read_csv(data)
df.head()
```

```python
# Add tops and bottoms flags
df['Tops'] = tops(df.High)
df['Bottoms'] = bottoms(df.Low)
df['Local time'] = df['Local time'].apply(lambda r: datetime.strptime(r, '%d.%m.%Y %H:%M:%S.%f GMT%z').replace(tzinfo=None))
```

```python
df.head()
```

```python
# We want to visualize the data.
# Since there are a lot of points and the plot can be very wide, we experiment with plotting 10 days data.
part_df = df[ (df['Local time'] > datetime(2018, 1, 1) ) & (df['Local time'] < datetime(2018, 1, 20) )]
```

```python
%matplotlib inline
def plot_data(df):
    df.plot(x='Local time', y='High', marker='o', markevery=find_max_indices(df.High), figsize = (18, 10))
    df.plot(x='Local time', y='Low', marker='o', markevery=find_min_indices(df.Low), figsize = (18, 10))
    plt.show()
plot_data(part_df)
```
